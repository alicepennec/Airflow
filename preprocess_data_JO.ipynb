```python
import pandas as pd
```

```python
path = 'data/fact_resultats_epreuves.csv'
df = pd.read_csv(path, sep=',')
```

```python
df.head()
```

```python
# split the dataframe in fives random parts and save as csv 
for i in range(5):
    df_part = df.sample(frac=0.2, random_state=i)
    df_part.to_csv(f'data/fact_resultats_epreuves_part_{i}.csv', index=False)
    df = df.drop(df_part.index)
```
